```python
import pandas as pd

# for getting the DataFrame from input string
def returnDataFrame_fromString(df_string):

    colmn_names = df_string.split('\n')[2][1:-1].split('|')
    colmn_names = [i.strip() for i in colmn_names]

    values = [[] for i in range(len(colmn_names))]

    for row in df_string.split('\n')[4:-2]:

        co = 0
        for val in row[1:-1].split('|'):

            val = val.strip()
            if val.isnumeric():
                values[co].append(int(val))
            else:
                values[co].append(val)
            co += 1
    df_dict = {col:val for (col, val) in zip(colmn_names, values)}
    return pd.DataFrame(df_dict)

# for making dataframe from df_string use this function: df = returnDataFrame_fromString(df_string)
```

```python
# https://leetcode.com/problems/big-countries/description/
"""
Problem description:

A country is big if:

it has an area of at least three million (i.e., 3000000 km2), or
it has a population of at least twenty-five million (i.e., 25000000).
Find the name, population, and area of the big countries.

Return the result table in any order.

"""

df_string = """
+-------------+-----------+---------+------------+--------------+
| name        | continent | area    | population | gdp          |
+-------------+-----------+---------+------------+--------------+
| Afghanistan | Asia      | 652230  | 25500100   | 20343000000  |
| Albania     | Europe    | 28748   | 2831741    | 12960000000  |
| Algeria     | Africa    | 2381741 | 37100000   | 188681000000 |
| Andorra     | Europe    | 468     | 78115      | 3712000000   |
| Angola      | Africa    | 1246700 | 20609294   | 100990000000 |
+-------------+-----------+---------+------------+--------------+
"""

import pandas as pd

def big_countries(world: pd.DataFrame) -> pd.DataFrame:

    world = world.filter(["name", "population", "area"]) # only filter these columns
    return world[(world['area'] >= 3000000) | (world['population'] >= 25000000)]
```

```python
# https://leetcode.com/problems/article-views-i/description/
"""
Problem description:

Write a solution to find all the authors that viewed at least one of their own articles.
Return the result table sorted by id in ascending order.
"""

df_string = """
+------------+-----------+-----------+------------+
| article_id | author_id | viewer_id | view_date  |
+------------+-----------+-----------+------------+
| 1          | 3         | 5         | 2019-08-01 |
| 1          | 3         | 6         | 2019-08-02 |
| 2          | 7         | 7         | 2019-08-01 |
| 2          | 7         | 6         | 2019-08-02 |
| 4          | 7         | 1         | 2019-07-22 |
| 3          | 4         | 4         | 2019-07-21 |
| 3          | 4         | 4         | 2019-07-21 |
+------------+-----------+-----------+------------+
"""

import pandas as pd

def article_views(views: pd.DataFrame) -> pd.DataFrame:

    unique_authors = views[views['author_id'] == views['viewer_id']]['author_id'].unique()
    unique_authors = sorted(unique_authors)
    result_df = pd.DataFrame({'id': unique_authors})
    return result_df
```

```python
# https://leetcode.com/problems/calculate-special-bonus/description/
"""
Problem description:

tag employees bounus:
employee will get bonus if their name doesn't start with M and their employee_id is odd
"""

df_string = """
+-------------+---------+--------+
| employee_id | name    | salary |
+-------------+---------+--------+
| 2           | Meir    | 3000   |
| 3           | Michael | 3800   |
| 7           | Addilyn | 7400   |
| 8           | Juan    | 6100   |
| 9           | Kannon  | 7700   |
+-------------+---------+--------+
"""

import pandas as pd

def calculate_special_bonus(employees: pd.DataFrame) -> pd.DataFrame:

    bonus = []
    employees.sort_values(by=['employee_id'], inplace=True)
    
    for lo in range(len(employees)):

        if employees.iloc[lo]['name'][0].lower() != 'm' and (employees.iloc[lo]['employee_id'] % 2 != 0):
            bonus.append(employees.iloc[lo]['salary'])
            
        else:
            bonus.append(0)
            
    df = pd.DataFrame({'employee_id':employees['employee_id'], 'bonus':bonus})
    return df

# with functions from solution
import pandas as pd

def calculate_special_bonus(employees: pd.DataFrame) -> pd.DataFrame:
    employees["bonus"] = employees.apply(lambda x : x["salary"] if x["employee_id"] % 2 != 0 and x["name"][0] != "M" else 0, axis = 1)
    return employees.loc[:,["employee_id", "bonus"]].sort_values(by=["employee_id"])
```

```python
# https://leetcode.com/problems/nth-highest-salary/description/
"""
Problem description:

find the N'th largest salary from the dataframe 
"""

df_string = """
+----+--------+
| id | salary |
+----+--------+
| 1  | 100    |
| 2  | 200    |
| 3  | 300    |
+----+--------+
"""

import pandas as pd

def nth_highest_salary(employee: pd.DataFrame, N: int) -> pd.DataFrame:

    # Drop any duplicate salary values to avoid counting duplicates as separate salary ranks
    unique_salaries = employee['salary'].drop_duplicates()

    # Sort the unique salaries in descending order and get the Nth highest salary
    sorted_salaries = unique_salaries.sort_values(ascending=False)

    # If N exceeds the number of unique salaries, return None
    if N > len(sorted_salaries):
        return pd.DataFrame({'Nth Highest Salary': [None]})
    
    # Get the Nth highest salary from the sorted salaries
    nth_highest = sorted_salaries.iloc[N - 1]
    
    return pd.DataFrame({'Nth Highest Salary': [nth_highest]})
```

```python
# https://leetcode.com/problems/department-highest-salary/description/
"""
Problem description:

find the employees with the highest salary from department dataframe
"""

employee = """
+----+-------+--------+--------------+
| id | name  | salary | departmentId |
+----+-------+--------+--------------+
| 1  | Joe   | 50000  | 1            |
| 2  | Jim   | 90000  | 1            |
| 3  | Henry | 80000  | 2            |
| 4  | Sam   | 60000  | 2            |
| 5  | Max   | 90000  | 1            |
| 6  | Max   | 90000  | 3            |
+----+-------+--------+--------------+
"""

department = """
+----+-------+
| id | name  |
+----+-------+
| 1  | IT    |
| 2  | Sales |
| 3  | pimp  |
+----+-------+
"""

import pandas as pd

def department_highest_salary(employee: pd.DataFrame, department: pd.DataFrame) -> pd.DataFrame:
    
    # merge function merge two dataframes right and left, attribute
    merged_df = employee.merge(department, left_on="departmentId", right_on="id", how="inner", suffixes=('_emp', '_dep'))
    # can use apply on groupby function
    merged_df = merged_df.groupby('departmentId').apply(lambda x: x[x['salary'] == x['salary'].max()])
    # to get flatten df
    merged_df = merged_df.reset_index(drop=True)
    
    merged_df = merged_df[['name_dep','name_emp','salary']].rename(columns={'name_dep':'Department', 'name_emp':'Employee', 'salary':'Salary'})
    return merged_df
```

```python
# https://leetcode.com/problems/rank-scores/description/
"""
Problem description:

sort the scores in decending order and tag their ranks
"""

scores = """
+----+-------+
| id | score |
+----+-------+
| 1  | 3.50  |
| 2  | 3.65  |
| 3  | 4.00  |
| 4  | 3.85  |
| 5  | 4.00  |
| 6  | 3.65  |
+----+-------+
"""


import pandas as pd

def order_scores(scores: pd.DataFrame) -> pd.DataFrame:

    # rank function documentation: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rank.html
    scores['rank'] = scores['score'].rank(method="dense", ascending=False)
    scores = scores.sort_values(by=['score'], ascending=False)
    # scores['rank'] = scores['rank'].apply(lambda x: int(x))

    return scores[['score', 'rank']]
```
